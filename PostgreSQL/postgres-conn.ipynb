```python
from sqlalchemy import create_engine
```

```python
import pandas as pd
import numpy as np
```

```python
all_statesdf = pd.read_csv("https://api.covidactnow.org/v2/states.timeseries.csv?apiKey=1540d77cf9a0422b88a2d90e989bbd23")
all_statesdf.head()
```

```python
new_df_1 = all_statesdf[["date","state","actuals.newCases",'actuals.newDeaths',"actuals.vaccinesDistributed","actuals.vaccinationsInitiated","actuals.vaccinationsCompleted",'metrics.vaccinationsInitiatedRatio',
       'metrics.vaccinationsCompletedRatio']].reset_index()

new_df_1 = new_df_1.rename(columns={"index": "id"})
new_df_1.set_index("id", inplace=True) 
# new_df_1.index = new_df_1.index.set_names(['id'])
new_df_1
```

```python
new_df_1.columns = ["date", "state", "newCases", "newDeaths", "vaccinesDistributed", "vaccinationsInitiated", "vaccinationsCompleted","vaccinationsInitiatedRatio","vaccinationsCompletedRatio"]
new_df_1
```

```python
connection_string = "postgres:postgres@localhost:5433/covid-post"
engine = create_engine(f'postgresql://{connection_string}')
```

```python
engine.table_names()
```

```python
new_df_1.to_sql(name='covidtimeseries', con=engine, if_exists='replace', index=True)
```
