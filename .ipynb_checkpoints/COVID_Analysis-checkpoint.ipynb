```python
import config
from config import api_key
import pandas as pd
import numpy as np
import requests
from pprint import pprint
import json
```

https://apidocs.covidactnow.org/api

```python
name_state = "CA"
```

```python
url = f"https://api.covidactnow.org/v2/state/{name_state}.timeseries.json?apiKey={api_key}"
```

```python
response = requests.get(url)
data = response.json()
print(json.dumps(data,indent=4,sort_keys=True))
```

```python
for i in data:
    print(i)
```

```python
parse_json_test = []
for i in data["actualsTimeseries"]:
    parse_json_test.append(i)
actualsTimeseries = pd.DataFrame(parse_json_test)
```
